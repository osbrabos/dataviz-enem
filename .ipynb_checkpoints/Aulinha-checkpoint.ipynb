# Importando as biblitoecas e os dados

```python
import pandas as pd #manipulaçao de dados
import matplotlib.pyplot as plt #visualizaçao
import seaborn as sns #visualizaçao
```

```python
plt.style.use('ggplot')
```

```python
df = pd.read_csv('train.csv') #importar os dados
df.head() #mostrar os 5 primeirso da tabela
```

```python
df.shape 
```

## Lidando com valores nulos

```python
df.isnull().sum()/df.shape[0]*100 #quantos de nulo em cada coluna
```

```python
df = df.drop(columns=['Cabin', 'Embarked']) #remove a coluna cabins
```

```python
mean = df['Age'].mean() #encontra a media da coluna  age
print(mean)
```

```python
# df['Age'] = df['Age'].fillna(mean) #substitui os fvalores nulos da coluna age pela media dela
df = df.dropna()
```

```python
df.head()
```

```python
df.shape
```

## Criando novas colunas a partir de outras

```python
df['Fare per age'] = df['Fare']/df['Age'] 
df.head()
```

```python
fig, ax = plt.subplots(figsize=(10, 5))
sns.scatterplot(data=df.query("Fare < 100"), x='Age', y='Fare', hue='Pclass', palette='tab10')
plt.show()
```
